### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
try:
    import matplotlib.pyplot as plt
    import numpy as np
except ImportError as e:
    print(f"Error: {e}\nPlease install the required libraries using: pip install matplotlib numpy")
    exit()

# Generate data
x = np.linspace(0, 10, 100)
y = x**2

# Create plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='y = x^2', color='b', linestyle='-', linewidth=2)

# Customizations
plt.title('Square Function Plot')
plt.xlabel('X values')
plt.ylabel('Y values (X squared)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Show plot
plt.show()
```
